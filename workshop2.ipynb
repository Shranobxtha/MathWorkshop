1. Write a Python function that takes a list and returns a new list with distinct elements from the
first list.
Sample List : [1,2,3,3,3,3,4,5]
Unique List : [1, 2, 3, 4, 5]

```python
def distinct(list):
  x = []
  for i in list:
    if i not in x:
      x.append(i)
  return x

list = []
n = int(input("Enter number of terms: "))
for i in range(n):
  ele = int(input("Enter elements for list: "))
  list.append(ele)
a = distinct(list)
print(a)
```

3. Write a program to print twin primes less than 1000. If two consecutive odd numbers are
both prime then they are known as twin primes.

```python
prime = []
for i in range(2, 1000):
    for j in range(2, i):
        if i % j == 0:
            break
    else:
        prime.append(i)
for i in range(3, 998, 2):
    if i in prime and (i + 2) in prime:
        print(i, i + 2)
```

4. A positive integer is called a perfect number if it is equal to the sum of all of its divisors,
including 1 but excluding the number itself. For example, 6 = 1 + 2 + 3. Write a
function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors
(excluding n).

```python
def sumOfDivisors(n):
  sum = 0
  for i in range(1, n):
    if n % i == 0:
      sum += i
  return sum
n = int(input("Enter a number: "));
if n == sumOfDivisors(n):
  print(n,"is a perfect number.") #6,28,496,...
else:
  print(n,"is not a perfect number.")
```

7. Write a function that converts a decimal number to binary number.

```python
def dectobin(n):
  binary ='' # empty string to add rem
  if n == 0:
    return 0
  while n > 0:
    rem = n % 2 # if user enters 15 then rem stores 1
    binary = str(rem) + binary # rem is converted to string and updated in the front of binary string variable
    n = n // 2 # if user enters 15 then n updates to 7
  return binary
n = int(input("Enter a number: "))
a=dectobin(n)
print(a)
```

8. Write a function prodDigits() that inputs a number and returns the product of digits of that
number.

```python
def prodDigits(n):
  product = 1
  if n == 0: #incase user inputs 0 as number of choice
    return 0
  while n > 0:
    rem = n % 10 # if user enters 123 rem stores 3
    product *= rem
    n = n // 10 # if user enters 123 then n becomes 12
  return product
n = int(input("Enter a number: "))
a=prodDigits(n)
print(a)
```

9. Write a function that will take in an unknown number of arguments and multiply all of them
together and run the function for these 2 sets of numbers:
a. 1,2,3,4,5
b. 12,13,14

```python
def multiply(*args): #(*args) allows for unknown number of arguments
  product = 1
  for num in args:
    product *= num
  return product
setA = {1,2,3,4,5}
setB = {12,13,14}
print(multiply(*setA)) #(*setA) unpacks the set from({1,2,3,4,5}) to (1,2,3,4,5)
print(multiply(*setB)) #(*setB) same reason as above
```
