Task 1:
Create a scatter graph of 10 random x and y values between 0 and 50. Label the axis to show
x – Time taken in seconds
y – Distance reached
Title “Buzzer Game Results”
One you have a working program, add the following code to your program;
colors=["red","green","blue","yellow","pink","orange","purple","brown"
,"cyan","magenta"]
And add to your scatter() call the ‘c=colors’ argument like so;
plt.scatter(x, y, c=colors)
What happens?

```python
import matplotlib.pyplot as plt
import random
x = []
y = []
for i in range(0,10):
  x.append(random.randint(0,50))
  y.append(random.randint(0,50))
colors = ["red","green","blue","yellow","pink","orange","purple","brown" ,"cyan","magenta"]
plt.xlabel("Time Taken in Seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")
plt.scatter(x,y,c=colors)
plt.show()
```

Task 2: Following data shows the change in population of different animals from 2017 to 2022.
a. Create a pie chart for the population of each year
b. Can we show all the pie chart on the single pane? Write the code.
image.png

```python
import matplotlib.pyplot as plt
a = [8,150,80] #2017
b = [54,77,54] #2018
c = [93,32,100] #2019
d = [116,11,76] #2020
e = [137,6,93] #2021
f = [184,1,72] #2022
groups = ['Bears','Dolphins','Whales']
plt.title('2017')
plt.pie(a,labels=groups)
plt.show()
print()
plt.title('2018')
plt.pie(b,labels=groups)
plt.show()
print()
plt.title('2019')
plt.pie(c,labels=groups)
plt.show()
print()
plt.title('2020')
plt.pie(d,labels=groups)
plt.show()
print()
plt.title('2021')
plt.pie(e,labels=groups)
plt.show()
print()
plt.title('2022')
plt.pie(f,labels=groups)
plt.show()

print()
#we can show all pie chart in a single pane using subplot
plt.figure(figsize=(12,10))  # for bigger canvas
plt.subplot(2,3,1)  # 1st position
plt.title('2017')
plt.pie(a,labels=groups)

plt.subplot(2,3,2)  # 2nd position
plt.title('2018')
plt.pie(b,labels=groups)

plt.subplot(2,3,3)  # 3rd position
plt.title('2019')
plt.pie(c,labels=groups)

plt.subplot(2,3,4)  # 4th position
plt.title('2020')
plt.pie(d,labels=groups)

plt.subplot(2,3,5)  # 5th position
plt.title('2021')
plt.pie(e,labels=groups)

plt.subplot(2,3,6)  # 6th position
plt.title('2022')
plt.pie(f,labels=groups)

plt.subplots_adjust(wspace=0.4, hspace=0.2)

plt.show()
```

Task 3: Construct a pie chart to visually display the favorite fruits of the students in a class based on the
given data: Mango - 45; Orange - 30; Plum - 15; Pineapple - 30; Melon – 30

```python
import matplotlib.pyplot as plt
y=[45,30,15,30,30]
groups=['Mango','Orange','Plum','Pineapple','Melon']
plt.figure(figsize=(6,6))
plt.title('Favorite Fruits of the students in a Class')
plt.pie(y,labels=groups,autopct='%1.1f%%')
plt.show()
```

Task 4: Here is some data to use for the tasks.
Car color Number seen
Red 123
Silver 230
Blue 78
White 193
Green 12
a) Create a bar chart for the car data above.
b) Implement the code “ plt.bar(x, y, width = 0.3)” to plot the bar chart, what is the effect?
c) Implement the code “ plt.barh(x, y)” to plot the bar chart, what is the effect?
d) Create a list of colors and using the “ plt.bar(x, y, color = {you list here} ) “set the bar
chart colors to be the same as the car colors each bar represents.
e) Create a pie chart of the car data above and add a title.
f) Test out the following pie chart additional arguments / settings (“...” is for the normal
code you have already)
A myexplode = [0.3, 0, 0, 0]
plt.pie(... , explode = myexplode)
B myexplode = [0.3, 0, 0, 0]
plt.pie(... , explode = myexplode,shadow = True)
C plt.legend(title = "Car Colours:")
D mycolours = ["Magenta", "Green", "Cyan", "Red"]

```python
import matplotlib.pyplot as plt
x=['Red','Silver','Blue','White','Green']
y=[123,230,78,193,12]
#a
plt.bar(x,y)
plt.title("Cars with certain color")
plt.show()
print()

#b
plt.bar(x,y,width=0.3) #width of bar in bargraph is adjusted to 0.3 unit
plt.title("Cars with certain color")
plt.show()
print()

#c
plt.barh(x,y) #makes the bar graph from vertical to horizontal
plt.title("Cars with certain color")
plt.show()
print()

#d
plt.bar(x,y,color=x) #gives each bar the same color that it's representing
plt.title("Cars with certain color")
plt.show()
print()

#e
plt.pie(y,labels=x,autopct='%1.1f%%',colors=x) #gives pie chart for the above data
plt.title("Cars with certain color")
plt.show()
print()

#f
myexplode=[0.3,0,0,0,0]

#f,A
plt.pie(y,labels=x,autopct='%1.1f%%',explode=myexplode,colors=x) #takes out a slice from pie chart, in this case for color red
plt.title("Cars with certain color")
plt.show()
print()

#f,B
plt.pie(y,labels=x,autopct='%1.1f%%',explode=myexplode,shadow=True,colors=x) #takes out a slice from pie chart, in this case for color red, and also gives shadow to pie chart
plt.title("Cars with certain color")
plt.show()
print()

#f,C
plt.pie(y,labels=x,autopct='%1.1f%%',explode=myexplode,shadow=True,colors=x) #adds legend to abovee graph
plt.title("Cars with certain color")
plt.legend(title='Car Colors')
plt.show()
print()

#f,D
mycolors=["Magenta","Green","Cyan","Red","Black"]
plt.pie(y,labels=x,autopct='%1.1f%%',explode=myexplode,shadow=True,colors=mycolors) #changes color of pie chart to colors from the list mycolors
plt.title("Cars with certain color")
plt.legend(title='Car Colors')
plt.show()
```
